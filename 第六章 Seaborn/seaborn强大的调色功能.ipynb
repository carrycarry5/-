```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def func_plot():
    x = np.linspace(0,14,100)
    for i in range(4):
        plt.plot(x, np.sin(x+i)*(i+0.75), label='np.sin(x+%s)*(%s+0.75)'%(i,i))
    plt.legend()
```

```python
func_plot()      # 导入seaborn之前
```
