```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

### distplot   同时画直方图和密度图

```python
s = pd.Series(np.random.randn(1000))
```

```python
sns.distplot(s)
```

```python
sns.distplot(s, hist=False)
```

```python
sns.distplot(s, kde=False, bins=20)
```

```python
sns.distplot(s,rug=True)
```

### kdeplot 画密度图

```python
sns.kdeplot(s, shade = True)          # 阴影
```

### rug密度图

```python
sns.rugplot(s)
```
