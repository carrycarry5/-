```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### axes_style 和 set_style     背景风格

```python
x = np.linspace(0,14,100)
y1 = np.sin(x)
y2 = np.sin(x+2)*1.25
```

```python
def func_plot():
    plt.plot(x,y1)   
    plt.plot(x,y2)
```

```python
func_plot()  # 没有引入seaborn之前
```

```python
import seaborn as sns
```

```python
func_plot()
```

### sns内置了5种风格

```python
style= ['darkgrid','dark','white','whitegrid','ticks']
```

```python
sns.set_style('dark')
```

```python
func_plot()
```

```python
sns.set_style('white')  # 默认
```

```python
func_plot()
```

```python
sns.set_style('ticks')
```

```python
func_plot()
```

```python
sns.set_style('darkgrid')
```

```python
sns.axes_style()        # 显示当前主题的参数 可以进行微调
```

```python
sns.set_style('darkgrid',{'grid.color': 'black'})   # 改成黑色网格
```

```python
func_plot()
```
