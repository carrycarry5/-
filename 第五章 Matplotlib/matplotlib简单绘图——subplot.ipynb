```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [1,2,3]
y1 = [1,2,3]
y2=[2,3,4]
y3 = [3,2,1]
y4 = [4,3,2]
```

```python
plt.subplot(2,2,1)   # 切分成两行两列，第三个参数表示的是第一个子图
plt.plot(x,y1 , 'b--')
plt.subplot(2,2,2)
plt.plot(x, y2, 'r--')
plt.subplot(2,2,3)
plt.plot(x, y3, 'b--')
plt.subplot(2,2,4)
plt.plot(x,y4,'r--')
plt.show()
```

```python
# 简便写法：
figure , ax = plt.subplots(2,2)
ax[0][0].plot(x,y1)
ax[0][1].plot(x,y2)
```

```python
fig , ax = plt.subplots(2,2)
ax[0,0].plot(x,y1)
ax[1,0].plot(x,y2)
```
