```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
s1 = pd.Series(np.random.randn(1000)).cumsum()  #累加
```

```python
# s1.plot()
plt.plot(s1)    # 也可以这么写
```

```python
s2 = pd.Series([1,2,3])
s2.plot(kind='bar')
```

```python
s2.plot(kind='barh')
```

```python
s2.plot(kind='hist')
```

```python
s2.plot(kind='box')
```

```python
s1.plot(grid='True',label='s1', title='series')  # 显示方格
```

```python
s2 = pd.Series(np.random.randn(1000)).cumsum()    #同时显示两个图
s1.plot()
s2.plot()
plt.show()
```

```python
# 通过subplot来显示多个图
```

```python
fig , ax = plt.subplots(2,1)
ax[0].plot(s1)
ax[1].plot(s2)
```
