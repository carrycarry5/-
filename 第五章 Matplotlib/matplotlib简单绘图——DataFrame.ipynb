```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame(
    data = np.random.randint(low=1,high=10,size=(10,4)),
    columns = ['A','B','C','D']
)
df
```

```python
#  plt.plot(df)
df.plot(style="--")
```

```python
df.plot(kind='bar')
```

```python
df.plot(kind='bar',stacked=True)
```

```python
df.plot(kind='area')
```

```python
for i in df.index:
    a = df.iloc[i]
    a.plot(label=str(i))
plt.legend()
plt.show()
```

### 

```python
for each in df.columns:
    df[each].plot(label=each)
plt.legend()
plt.show()
```
