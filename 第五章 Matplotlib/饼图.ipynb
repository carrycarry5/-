```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
s = pd.Series(data=[1,2,3,4], index=['a','b','c','d'])
s.plot()
```

```python
plt.pie([1,2,3,4],labels=['a','b','c','d'])
plt.legend()
plt.show()
```

```python
plt.pie(s,labels=s.index)
plt.legend()
plt.show()
```
