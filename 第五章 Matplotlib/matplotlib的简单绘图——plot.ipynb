```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [1,2,3]
y = [4,5,6]
plt.plot(x,y)
```

```python
plt.plot(x,y,'bo')
```

```python
plt.plot(x,y,'r*')
```

```python
c = [7,8,9]
d = [10,11,12]
plt.plot(x,y,'r--', c,d,'g*')           # 一张图里面可以有多个函数
```

```python
t = np.arange(0.0,2.0,step=0.1)
t
```

```python
s = np.sin(t*np.pi)
s
```

```python
plt.plot(t,s, t*2,s)
```
