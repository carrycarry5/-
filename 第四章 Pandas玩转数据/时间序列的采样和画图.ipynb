```python
import numpy as np
import pandas as pd
```

```python
t_range = pd.date_range('2018-1-1','2018-12-31')
t_range
```

```python
s1 = pd.Series(data= np.random.randn(len(t_range)), index = t_range)
s1
```

```python
s1['2018-01']   # 返回1月份的数据
```

```python
s1['2018-01'].mean()  # 一月的平均值
```

```python
s1.resample('M').mean()  # 每月采样  
```
